```python
import pyodbc
import yfinance as yf
import pandas as pd

# Verbindungsstring
conn_str = (
    r'Driver=SQL Server;'
    r'Server=.\SQLEXPRESS;'
    r'Database=studienprojekt;'
    r'Trusted_Connection=yes;'
)

# Verbindung zur Datenbank herstellen
cnxn = pyodbc.connect(conn_str)
cursor = cnxn.cursor()

# Tabelle prüfen
table_name = "stock_data"  # Name der Zieltabelle
start_date = "2016-01-01"  # Startdatum, ab dem Daten überprüft werden sollen
end_date = "2021-10-01"  # Enddatum, bis zu dem Daten überprüft werden sollen

# Daten aus der Tabelle in einen DataFrame laden
select_query = f"SELECT * FROM {table_name} WHERE timestamp >= ? AND timestamp <= ?"
df_existing_data = pd.read_sql(select_query, cnxn, params=(start_date, end_date))

# Verbindung schließen
cursor.close()
cnxn.close()

# DataFrame anzeigen
print(df_existing_data.head())
```

```python
from statsmodels.tsa.stattools import adfuller

# Dickey-Fuller-Test durchführen
result = adfuller(df_existing_data["close"])

# Ergebnisse anzeigen
print("Dickey-Fuller-Test:")
print(f'Teststatistik: {result[0]}')
print(f'p-Wert: {result[1]}')
print('Kritische Werte:')
for key, value in result[4].items():
    print(f'\t{key}: {value}')
```

```python
print(df_existing_data.shape)
train=df_existing_data.iloc[:-30]
test=df_existing_data.iloc[-30:]
print(train.shape,test.shape)
```

```python
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(train['close'], order=(1, 1, 0))
model = model.fit()
print(model.summary())
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Vorhersage für den nächsten Monat
n_periods = 30  # Anzahl der Perioden (Tage) für die Vorhersage
forecast, conf_int = model.predict(n_periods=n_periods, return_conf_int=True)

# Erzeugen des Datumsbereichs für die Vorhersage
start_date = pd.to_datetime(df_existing_data["timestamp"].iloc[-1]) + pd.DateOffset(days=1)
end_date = start_date + pd.DateOffset(days=n_periods)
forecast_dates = pd.date_range(start=start_date, end=end_date, freq='D')

# Vorhersagen in einen DataFrame laden
forecast_df = pd.DataFrame({'Date': forecast_dates, 'Forecast': forecast})
forecast_df.set_index('Date', inplace=True)

# Plot der Vorhersage und des Konfidenzintervalls
plt.figure(figsize=(12, 6))
plt.plot(df_existing_data["timestamp"], df_existing_data["close"], label='Existing Data')
plt.plot(forecast_df.index, forecast_df['Forecast'], label='Forecast')
plt.fill_between(forecast_df.index, conf_int[:, 0], conf_int[:, 1], color='gray', alpha=0.2, label='Confidence Interval')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.title('ARIMA Forecast')
plt.legend()
plt.show()
```
